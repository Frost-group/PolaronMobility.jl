```julia
# JuliaBox-Example.ipynb - live notebook for polarons
# Jarvist Moore Frost and Bradley A A Martin

push!(LOAD_PATH,"./src/") # load module from local directory
using PolaronMobility, Unitful, Plots 
```

```julia
ϵ_optical = 4.5
ϵ_static = 24.1
electron_effective_mass =  0.12 # Electron mass me = 9.11e-31 kg
phonon_freq = 2.25              # Terahertz 
T = 300                         # Kelvin
MAPIe = material(ϵ_optical, ϵ_static, electron_effective_mass, phonon_freq)
MAPIe_polaron = frohlichpolaron(MAPIe, T)
addunits!(MAPIe_polaron)
display(MAPIe_polaron)
```

```julia
hole_effective_mass = 0.15 # electron masses me = 9.11e-31 kg
MAPIh = material(ϵ_optical, ϵ_static, hole_effective_mass, phonon_freq)
MAPIh_polaron = frohlichpolaron(MAPIh, T)
addunits!(MAPIh_polaron)
display(MAPIh_polaron)
```

```julia
MAPIe_polaron_energy = MAPIe_polaron.F .|> u"meV"
```

```julia
MAPIh_polaron_energy = MAPIh_polaron.F .|> u"meV"
```

```julia
MAPIe_mobility = MAPIe_polaron.μ |> u"cm^2/V/s"
```

```julia
MAPIh_mobility = MAPIh_polaron.μ |> u"cm^2/V/s"
```

```julia
# Generate Freq-dep mobility data; and plot
Trange = [0, 300]       # Kelvin
Ωrange = 0.0:0.2:10     # Terahertz
MAPIe_polaron = frohlichpolaron(MAPIe, Trange, Ωrange; verbose = true)
addunits!(MAPIe_polaron)
```

```julia
## Calculate real and imaginary conductivity
plot(MAPIe_polaron.Ω .|> u"THz2π", real.(MAPIe_polaron.σ) .|> u"μS", label = ["Reσ, T=0K" "Reσ, T=300K"], markersize = 3, marker = [:rect :diamond], xlab = "Frequency", ylab = "Polaron Conductivity", ylims = (0,400))
plot!(MAPIe_polaron.Ω .|> u"THz2π", imag.(MAPIe_polaron.σ) .|> u"μS", label = ["Imσ, T=0K" "Imσ, T=300K"], markersize = 3, marker = [:uptriangle :downtriangle])
```

```julia
## Calculate real and imaginary impedance
plot(MAPIe_polaron.Ω .|> u"THz2π", real.(MAPIe_polaron.z) .|> u"kΩ", label = ["Rez, T=0K" "Rez, T=300K"], markersize = 3, marker = [:rect :diamond], xlab = "Frequency", ylab = "Polaron Impedance", legend = :topright)
plot!(MAPIe_polaron.Ω .|> u"THz2π", imag.(MAPIe_polaron.z) .|> u"kΩ", label = ["Imz, T=0K" "Imz, T=300K"], markersize = 3, marker = [:uptriangle :downtriangle])
```

```julia
## Calculate real and imaginary cmemory function
plot(MAPIe_polaron.Ω .|> u"THz2π", pustrip.(real.(MAPIe_polaron.χ)), label = ["Reχ, T=0K" "Reχ, T=300K"], markersize = 3, marker = [:rect :diamond], xlab = "Frequency", ylab = "Polaron Memory Function (arb)", legend = :right)
plot!(MAPIe_polaron.Ω .|> u"THz2π", pustrip.(imag.(MAPIe_polaron.χ)), label = ["Imχ, T=0K" "Imχ, T=300K"], markersize = 3, marker = [:uptriangle :downtriangle])
```
